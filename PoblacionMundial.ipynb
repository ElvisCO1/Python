```python
import re

def names():
    simple_string = """Amy is 5 years old, and her sister Mary is 2 years old. 
    Ruth and Peter, their parents, have 3 kids."""
    
    # Patrón para encontrar palabras que comiencen con una letra mayúscula seguida de letras minúsculas
    name_pattern = r'\b[A-Z][a-z]+\b'

    # Utiliza re.findall() para encontrar todas las coincidencias del patrón en la cadena
    names_found = re.findall(name_pattern, simple_string)
    
    # Devuelve la lista de nombres encontrados
    return names_found

# Llama a la función names() y almacena el resultado en una variable
found_names = names()

# Imprime los nombres encontrados
print(found_names)
```
